```python
Question 4- Check how many times a given number can be divided by 3 before it is less than or equal to 10.

#Answer 4

number = int(input("Enter a number: "))
count = 0

while number > 10:
    number = number / 3
    count += 1

print("The number can be divided by 3", count, "times before it is less than or equal to 10.")
```

```python
Question 5- Why and when to use while loop in python give a detailed description with example.

#Answer 5

A 'while'loop is a type of loop in Python that repeatedly executes a block of code as long as a 
condition is true. The basic syntax for a while loop in Python is:
    
    while condition:
    # code to execute

Here, 'condition' is an expression that evaluates to either True or False. The code inside the loop
is executed repeatedly as long as the condition is True. When the condition becomes False, the loop
terminates, and execution continues with the next statement after the loop.
    
The 'while' loop is useful when you want to repeatedly execute a block of code as long as a certain 
condition is met. Some examples of when to use a 'while' loop in Python include:
    
Repeating a process until a certain condition is met: Suppose you want to repeatedly ask the user
for input until they enter a valid value. You can use a' while' loop to keep asking for input until the
user enters a valid value.
    
    
    valid_input = False

while not valid_input:
    user_input = input("Enter a number: ")
    if user_input.isdigit():
        valid_input = True
        number = int(user_input)
    else:
        print("Invalid input. Please enter a number.")
        
        

Iterating over a 'list' or sequence: Suppose you want to iterate over a list or sequence and perform some
operation on each element until a certain condition is met. You can use a 'while' loop to iterate over the
list until the condition is met.

numbers = [1, 2, 3, 4, 5]
i = 0

while i < len(numbers):
    print(numbers[i])
    i += 1
```

```python
Question 7-Reverse a while loop to display the number from 10 to 1.


#Answer 7

num = 10
while num >= 1:
    print(num)
    num -= 1

In this code, we start with the variable num set to 10. We use a 'while' loop with the condition
num >= 1 to iterate 'while' num is greater than or equal to 1. Inside the loop, we print the value of 
num 'and' then decrement it by 1 with the num -= 1 statement. 'This' causes the loop to iterate backwards,
printing the numbers from 10 to 1. 
```
